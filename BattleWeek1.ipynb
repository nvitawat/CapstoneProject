#### 2.1 Borough, Neighborhood and Coordinates(latitude/longitude)
Borough, Neighborhood and Coordinates(latitude/longitude) of Bangkok, got from here https://data.go.th/Default.aspx. It provided latitude/longitude data of city in Thailand.

```python
import pandas as pd 
DF = pd.read_excel('data/TAMBON.xlsx')
DF.head(5)
```

##### Understand the data set: Each column is represented 
      - TA_ID       is Neighborhood ID
      - TAMBON_E    is Neighborhood Name in English
      - TAMBON_T    is Neighborhood Name in Thai
      - AM_ID       is Borough ID
      - AMPHOE_E    is Borough Name in English
      - AMPHOE_T    is Borough Name in Thai
      - CH_ID       is City ID
      - CHANGWAT_E  is CITY Name in English
      - CHANGWAT_T  is CITY Namein Thai
      
We will prepare data set follow these
1. Select 'Bangkok' 
2. Select column 'TA_ID', 'TAMBON_E','AMPHOE_E', 'LAT', 'LONG'
3. Change column_name 
      - TA_ID      to NBH_ID
      - TAMBON_E   to Neigborhood
      - AMPHOE_E   to Borough
      - LAT        to latitude
      - LONG       to longitude 

```python
# Select 'Bangkok'
BKKDF = DF[DF['CHANGWAT_E'] == 'Bangkok']
# Select column
BKKDF = BKKDF[['TA_ID', 'AMPHOE_E','TAMBON_E', 'LAT','LONG']].copy()
# Change column_name
columns = ['NBH_ID', 'Borough', 'Neighborhood',  'Latitude', 'Longitude']
BKKDF.columns = [columns]                                       
BKKDF.head(5)
```

```python
print ('Nunber of Borough = ', BKKDF['Borough'].nunique())
print ('Number of Neighborhood = ', BKKDF['NBH_ID'].nunique())    
```
